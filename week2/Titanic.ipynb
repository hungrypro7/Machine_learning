다음 명령으로 타이타닉호의 승객 데이터를 데이터프레임으로 읽어올 수 있음. 이 명령을 실행하려면 seaborn 패키지가 설치되어 있어야 한다.

- Q4, 타이타닉호 승객중 1등실 선실의 여성 승객의 평균 나이를 구하라.

```python
# Q1. sort_values 메서드를 사용하여 타이타닉호 승객에 대해 성별(sex) 인원수, 나이별(age) 인원수, 선실별(class) 인원수, 사망/생존(alive) 인원수를 구하라.
titanic['sex'].sort_values().value_counts()
```

```python
titanic['class'].value_counts()
```

```python
titanic['alive'].value_counts()
```

```python
# Q2. 타이타닉호 승객의 평균 나이를 구하라.
titanic['age'].mean()
```

```python
# Q3. 타이타닉호 승객중 여성 승객의 평균 나이를 구하라.
titanic.loc[titanic.sex == 'female']['age'].mean()
```

```python
# Q4, 타이타닉호 승객중 1등실 선실의 여성 승객의 평균 나이를 구하라.
titanic.loc[(titanic['class'] == 'First') & (titanic.sex == 'female')]['age'].mean()
```

다음과 같이 타이타닉호의 승객 중 나이 20살을 기준으로 성인(adult)과 미성년자(child)를 구별하는 라벨 열을 만들 수 있다.

```python
titanic["adult/child"] = titanic.apply(lambda r: "adult" if r.age >= 20 else "child", axis=1)
titanic.tail()
```

- Q5. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만들어라. category1 카테고리는 다음과 같이 정의하라.

    20살이 넘으면 성별을 그대로 사용한다.

    20살 미만이면 성별에 관계없이 “child”라고 한다.
    
- Q6. 타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고쳐라.

- Q7. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category2 열을 만들어라. category2 카테고리는 다음과 같이 정의된다.

    성별을 나타내는 문자열 male 또는 female로 시작한다.

    성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열이 온다.

    예를 들어 27살 남성은 male27 값이 된다.

```python
# Q5. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만들어라. category1 카테고리는 다음과 같이 정의하라.
titanic["adult/child"] = titanic.apply(lambda r: r['sex'] if r.age >= 20 else "child", axis=1)
titanic
```

```python
# Q6. 타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고쳐라.
age_mean = titanic.age.mean()
age_mean = round(age_mean)
print(age_mean)

import pandas as pd
titanic['age'] = titanic.apply(lambda x: age_mean if pd.isna(x.age) else x.age, axis=1).astype('i')
titanic.head(30)
```

```python
# Q7. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category2 열을 만들어라. category2 카테고리는 다음과 같이 정의된다.
titanic['category2'] = titanic.sex+titanic.age.astype(str)
titanic
```
